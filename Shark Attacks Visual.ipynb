```python
# IMPORT THE NEW DATAFRAME, POST CLEANING

shark = pd.read_csv("data/shark_attack_clean.csv",encoding = "ISO-8859-1")
```

```python
# CREATE A NEW DATA FRAME W/O "Unknown" TO PLOT THE SIX CATEGORIES OF HOUR

shark2 = shark.loc[shark["Time_merged"] != 'Unknown']
```

```python
shark7 = shark2.loc[shark2["Fatal (Y/N)"] != 'UNKNOWN']
```

```python
# CREATE THE TABLE THAT WE WANT TO LATTER PLOT

hour_attack = px.histogram(shark7, x="Time_merged",color="Fatal (Y/N)",title="ATTACKS BY HOUR")
```

```python
# FIRST PLOT TO CONTRAST HYPOTHESES NUMBER 1

hour_attack.show()
```
