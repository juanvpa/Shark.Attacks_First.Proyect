```python
# IMPORT LIBRERY

import pandas as pd
import numpy as np
import re
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
# IMPORT THE NEW DATAFRAME AS CSV, POST CLEANING

shark = pd.read_csv("data/shark_attack_clean.csv",encoding = "ISO-8859-1")
```

```python
# HYPOTHESES I
```

```python
# CREATE A NEW DATA FRAME W/O "Unknown" TO PLOT THE SIX CATEGORIES OF HOUR

shark2 = shark.loc[shark["Time_merged"] != 'Unknown']
```

```python
shark7 = shark2.loc[shark2["Fatal (Y/N)"] != 'UNKNOWN']
```

```python
# CREATE THE TABLE THAT WE WANT TO LATTER PLOT

hour_attack = px.histogram(shark7, x="Time_merged",color="Fatal (Y/N)",title="ATTACKS BY HOUR")
```

```python
# FIRST PLOT TO CONTRAST HYPOTHESES NUMBER 1

hour_attack.show()
```

```python
# 1. Highest % of attacks occurs during morning & afternoon
# Actually, the highest amount of attacks are during "Noon", then "Afternoon & then "Morning"
# The highest % of fatality is during "Night"
# % Fatality (Afternoon 15% | Noon 18% | Sunrise 17% | Morning 16% | Evening 26% | Night 39%)
```
